```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def saltPepperNoise(image):
  row,col = image.shape
  s_vs_p = 0.5
  amount = 0.04
  noisy = np.copy(image)
  num_salt = int(np.ceil(amount*image.size*s_vs_p))
  corrds = [np.random.randint(0,i-1,num_salt) for i in image.shape]
  noisy[corrds] = 1
  num_pep = int(np.ceil(amount*image.size*s_vs_p))
  corrds = [np.random.randint(0,i-1,num_pep) for i in image.shape]
  noisy[corrds] = 0
  return noisy
imgpath = '/media/sunflower.png'
img = cv2.imread(imgpath, cv2.CASCADE_SCALE_IMAGE)
img = img/255
noise_img = saltPepperNoise(img)

plt.figure(figsize=(8,8))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')

plt.subplot(2, 2, 2)
plt.imshow(noise_img, cmap='gray')

plt.subplot(2, 2, 3)
histogram, bin_edges = np.histogram(img, bins=256, range=(0, 1))
plt.plot(bin_edges[0:-1], histogram)

plt.subplot(2, 2, 4)
histogram, bin_edges = np.histogram(noise_img, bins=256, range=(0, 1))
plt.plot(bin_edges[0:-1], histogram)

```
