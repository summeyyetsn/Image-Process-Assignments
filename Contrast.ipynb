```python
import cv2
import matplotlib.pyplot as plt
imgpath = '/media/handwriting.jpeg'
img = cv2.imread(imgpath, cv2.CASCADE_SCALE_IMAGE)
gray = cv2.imread(imgpath, 0)
clahe = cv2.createCLAHE(clipLimit =7)
cl_img = clahe.apply(img)


plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')

plt.subplot(2, 2, 2)
plt.imshow(cl_img, cmap='gray')

plt.subplot(2, 2, 3)
plt.hist(img.flat, bins=100, range=(100, 255))

plt.subplot(2, 2, 4)
plt.hist(cl_img.flat, bins=100, range=(100, 255))


plt.show()
```
