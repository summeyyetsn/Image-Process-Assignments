```python
import cv2
import matplotlib.pyplot as plt
imgpath = '/media/sunflower.png'
img = cv2.imread(imgpath)
gray = cv2.imread(imgpath, 0)
colored_negative = abs(255-img)
gray_negative = abs(255-gray)

imgs = [ gray, gray_negative]
title = [ 'gray','gray-negative']


plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title(title[1])
plt.imshow(imgs[1], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```
