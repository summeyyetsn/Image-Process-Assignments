```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def gammaCorrection(src, gamma):
    gamma_new = np.array(255*(src/255)**gamma, dtype= 'uint8')
    return gamma_new

img = cv2.imread('/media/sunflower.png',cv2.IMREAD_GRAYSCALE)
j=1
plt.subplot(2, 2, j), plt.imshow(img, cmap='gray')
for i in [3.0,4.0,5.0]:
  j=j+1
  gammaImg = gammaCorrection(img, i)
  plt.subplot(2, 2, j)
  plt.imshow(gammaImg, cmap='gray')
plt.show()
```
